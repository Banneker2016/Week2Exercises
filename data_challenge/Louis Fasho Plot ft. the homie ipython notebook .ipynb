```python
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib notebook

time,data,error=np.loadtxt('data.txt',unpack=True)
mask= np.where(data>-500 & ~np.isnan(data))
time=time[mask]
data=data[mask]
error=error[mask]
plt.plot(time,data)
plt.xlabel('Time(years)')
plt.ylabel('Global Temp(C)')
mask 
```

```python
from scipy.optimize import curve_fit
def func(x,a,b):
    return a*x+b

popt, pcov=curve_fit(func,time,data,sigma=error)

print popt 
```

```python
%matplotlib notebook
model=func(time,popt[0],popt[1])
plt.errorbar(time,data,yerr=error)
plt.plot(time,model)
residual=data-model
plt.xlabel('Time(s)')
plt.ylabel('Global Temp(C)')
plt.legend
```
